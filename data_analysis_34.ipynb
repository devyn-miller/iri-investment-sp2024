```python
import pandas as pd

# Load the CSV files
df1 = pd.read_csv('investment_2023-12-11.csv')
df2 = pd.read_csv('investment_S2_2023-12-13.csv')
```

```python
df1.describe()
```

```python
df2.describe()
```

```python
# Combine the two datasets
combined_df = pd.concat([df1, df2])

# Basic information about the dataset
combined_df.describe()
```

```python
# Example analysis: Count of unique subjects
unique_subjects = combined_df['SubjectID'].nunique()
print(f'Unique Subjects: {unique_subjects}')

# Example analysis: Average parts completed per job
average_parts_completed = combined_df['Parts Completed'].mean()
print(f'Average Parts Completed: {average_parts_completed}')

# Example analysis: Acceptance rate
acceptance_rate = combined_df[combined_df['Action Type'] == 'Offer']['Accepted/Worked On'].mean()
print(f'Acceptance Rate: {acceptance_rate}')

# Example analysis: Payment received distribution
payment_received = combined_df['Payment Received'].dropna().astype(float)  # Ensure it's float
print(payment_received.describe())
```

```python
import statsmodels.api as sm

# Ensure the data is clean and ready for analysis
combined_df = combined_df.dropna(subset=['Parts Completed', 'Payment Received'])
X = combined_df['Parts Completed']
y = combined_df['Payment Received']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print out the statistics
print(model.summary())
```
